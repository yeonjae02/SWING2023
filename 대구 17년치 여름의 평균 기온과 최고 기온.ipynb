```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows, Mac일 때는 AppleGothic
#matplotlib.rcParams['font.family'] = 'HYGungSo-Bold'# 궁서체
matplotlib.rcParams['font.size'] = 15 # 폰트 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시 마이너스 글자가 깨지는 것방지
```

```python
df = pd.read_csv("대구 17년치 최고 기온 데이터.csv", engine = "python",encoding = "CP949")
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.title("대구 17년치 여름의 평균 기온과 최고 기온")
plt.plot(df['연도'], df['평균기온(℃)'], label="평균기온(℃)")
plt.plot(df['연도'], df['최고기온(℃)'], label="최고기온(℃)")
plt.grid()
plt.legend(ncol=2)
plt.savefig('대구 17년치 평균 기온.png', dpi=300)
```
